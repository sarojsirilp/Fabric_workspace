```python
def compare_schemas(schema1, schema2):
  if len(schema1.fields) != len(schema2.fields):
    print("Schemas differ in number of fields")
    return False

  # Loop through each field and compare details
  for i in range(len(schema1.fields)):
    field1 = schema1.fields[i]
    field2 = schema2.fields[i]

    # Compare field name
    if field1.name != field2.name:
      print(f"Field name mismatch: {field1.name} vs {field2.name}")
      return False

    # Compare data type
    if field1.dataType != field2.dataType:
      print(f"Data type mismatch: {field1.dataType} vs {field2.dataType}")
      return False

    # Compare nullability
    if field1.nullable != field2.nullable:
      print(f"Nullability mismatch: {field1.nullable} vs {field2.nullable}")
      return False

  # If all fields match, schemas are identical
  print("Schemas are identical")
  return True
```

```python
compare_schemas(schema1, schema2)
```

```python
def compare_schemas(schema1, schema2):
  mismatches = []

  # Check if number of fields differ
  if len(schema1.fields) != len(schema2.fields):
    mismatches.append(f"Schemas differ in number of fields: {len(schema1.fields)} vs {len(schema2.fields)}")

  # Loop through each field and compare details
  for i in range(len(schema1.fields)):
    field1 = schema1.fields[i]
    field2 = schema2.fields[i]

    # Compare field name
    if field1.name != field2.name:
      mismatches.append(f"Field name mismatch: {field1.name} vs {field2.name}")

    # Compare data type
    if field1.dataType != field2.dataType:
      mismatches.append(f"Data type mismatch for field '{field1.name}': {field1.dataType} vs {field2.dataType}")

    # Compare nullability
    if field1.nullable != field2.nullable:
      mismatches.append(f"Nullability mismatch for field '{field1.name}': {field1.nullable} vs {field2.nullable}")
  return mismatches

mismatches = compare_schemas(schema1, schema2)

if mismatches:
  print("Following schema mismatches were found:")
  for mismatch in mismatches:
    print(mismatch)
else:
  print("Schemas are identical")
```

```python
from tabulate import tabulate

comparison_results = []

# Iterate over the fields and compare names
for i in range(len(schema1.fields)):
    field1 = schema1.fields[i]
    field2 = schema2.fields[i]

    # Check if the field names match
    if field1.name == field2.name:
        match_status = "Match"
    else:
        match_status = "Mismatch"

    # Append the comparison result
    comparison_results.append([field1.name, field2.name, match_status])

# Define headers for the table
headers = ["Schema1 Field Name", "Schema2 Field Name", "Match Status"]

# Print the comparison results in tabular form
print(tabulate(comparison_results, headers=headers, tablefmt="grid"))
```
